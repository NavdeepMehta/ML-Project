```python
def impute_income(cols):
    Education = cols[0]
    Income = cols[1]
    if pd.isnull(Income):
        if Education == 'Graduation':
            return x['Income'][2]
        elif Education == 'PhD':
            return x['Income'][4]
        elif Education == 'Master':
            return x['Income'][3]
        elif Education == '2n Cycle':
            return x['Income'][0]
        else :
            return x['Income'][1]
    else:
        return Income
        
```
